# Functions of activations 

```python
def ReLU(x):
    if x < 0:
        return 0
    elif x >= 0:
        return x
 
```

```python
def ReLU(x):
    if x < 0:
        return 0
    elif x >= 0:
        return x

def der_ReLU(x):
    if x < 0:
        return 0 
    elif x >= 0:
        return 1 

x = [i for i in np.arange(-10, 10, 0.25).tolist() ]
y = [ReLU(i) for i in x]
der = [der_ReLU(i) for i in x]
plt.plot(x,y, label = "ReLU")
plt.plot(x, der, label = "Derevative")
plt.title("ReLU")
plt.legend()
plt.grid()
plt.show()
```

```python
def ReLU6(x):
    return min(max(0,x),6)
def der_ReLU6(x):
    if x <= 0:
        return 0
    elif 0 <= x <= 6:
        return 1
    else: return 0
x = [i for i in np.arange(-10, 10, 0.25).tolist()]
y = [ReLU6(i) for i in x]   
der = [der_ReLU6(i) for i in x]
plt.plot(x,y, label = "ReLU6")
plt.plot(x, der, label = "Derevative")
plt.grid()
plt.title("ReLU6")
plt.legend()
plt.show()
```

```python
def ELU(x):
    if x > 0:
        return x
    else: return (np.exp(x) - 1)

def der_ELU(x):
    if x > 0:
        return 1
    else: return np.exp(x)

x = [i for i in np.arange(-10, 10, 0.25).tolist()]
y = [ELU(i) for i in x]   
der = [der_ELU(i) for i in x]
plt.plot(x,y, label = "ELU")
plt.plot(x, der, label = "Derevative")
plt.grid()
plt.title("ELU")
plt.legend()
plt.show()
```

```python
def Softplus(x):
     return np.log(1 + np.exp(x))
def der_Softplus(x):
    return (np.exp(x))/(1 + np.exp(x))

x = [i for i in np.arange(-10, 10, 0.25).tolist()]
y = [Softplus(i) for i in x]   
der = [der_Softplus(i) for i in x]
plt.plot(x,y, label = "Softplus")
plt.plot(x, der, label = "Derevative")
plt.grid()
plt.title("Softplus")
plt.legend()
plt.show()
```

```python
def SiLU(x):
    return x/(1 + np.exp(-x))
    
def der_SiLU(x):
    return (1 + (1 + x) * np.exp(-x)   )/( 1 + np.exp(-x))**2
    
x = [i for i in np.arange(-10, 10, 0.25).tolist()]
y = [SiLU(i) for i in x] 
der = [der_SiLU(i) for i in x]
plt.plot(x,y, label = "SiLU")
plt.plot(x, der, label = "Derevative")
plt.grid()
plt.title("SiLU")
plt.legend()
plt.show()
plt.show()
```

# Losses

```python
def bce(y_pred, y):
    epsilon = 1e-15  # to prevent division by zero
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)  
    bce = - (y * np.log(y_pred) + (1 - y) * np.log(1- y_pred))
    return  bce

```

```python
y_pred_range = np.linspace(0.01, 0.99, 100)
y_true = 1



bce_loss_values = bce(y_pred_range,y_true)
plt.plot(y_pred_range, bce_loss_values)
plt.title('Binary Cross-Entropy Loss Function')
plt.xlabel('Predicted Probability')
plt.ylabel('Loss')
plt.grid(True)
plt.show()
```

```python
def KLdiv(p,q):
    # Relative entropy 
    #p - real value 
    #q - predicted value 
    epsilon = 1e-15  # to prevent division by zero
    p = np.clip(p, epsilon, 1 - epsilon) # clip to prevent log(0)
    q = np.clip(q, epsilon, 1 - epsilon)
    d = np.sum( p * np.log(p/q))
    return d
    
```

```python
x = np.array([1,1,0,0])
y = np.array([0.5,0.6,0.8,0.1])
KLdiv(x,y)
```

```python
def l1_loss(y, y_pred):
    return np.abs(y - y_pred)

y_true_values = np.linspace(-5, 5, 100)
y = 0

l1_loss_values = l1_loss(y_true_values, y)

plt.plot(y_true_values, l1_loss_values)
plt.title('L1 Loss Function')
plt.xlabel('True Values')
plt.ylabel('L1 Loss')
plt.grid()
plt.show()
```
